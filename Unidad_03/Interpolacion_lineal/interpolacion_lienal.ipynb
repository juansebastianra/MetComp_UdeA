```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
datos=np.loadtxt("datos.dat")
x_list=[]
y_list=[]
for i in range(0,datos.shape[0]):
    x_list.append(datos[i][0])
    y_list.append(datos[i][1])
    
#print(datos.shape)
#print(x_list,"\n",y_list)
def Interpoladorlienal(x_list,y_list,x):
    for i in range(0,len(x_list)-1):
        if(x>=x_list[i] and x<=x_list[i+1]):
            x_prev=x_list[i]
            y_prev=y_list[i] 
            x_next= x_list[i+1]
            y_next= y_list[i+1]
            y=(y_next-y_prev)/(x_next-x_prev)*(x-x_prev)+y_prev
            return y
x_interpol=np.linspace(min(x_list),max(x_list),1000)
y_interpol=[]
for i in range(0,len(x_interpol)):
    y_interpol.append(Interpoladorlienal(x_list,y_list,x_interpol[i]))
#print(x_interpol)
#print(Interpoladorlienal(x_list,y_list,7.1))         
```

```python
fig=plt.figure()
ax=plt.axes()
ax.plot(x_list,y_list,"*",markersize="10",label="datos discretos")
ax.plot(x_interpol,y_interpol,".",label="puntos interpolados")
ax.plot(x_interpol,np.sin(x_interpol),label="sin(x)")
ax.legend()
plt.grid()
plt.show()
```
